*I create two variables: one that contains the "numeric" part (e.g. 0, ... 9) and one that contains the "character" part (e.g. M, G, k, ..)*

```python
df['num_size'] = df['Size'].str.extract('([0-9.]*[0-9]+)', expand=False).str.strip() 
df['type_size'] = df['Size'].str[-1]
```

```python
df.head()
```

```python
df['type_size'].value_counts()
```

*Let's examine what "e" and "+" represent*

```python
df_e = df[df['type_size'] == 'e']
df_e['Size'].value_counts()
```

*For these records the variable does not return a correct value, therefore it is rightly managed as NaN*

```python
df_plus = df[df['type_size'] == '+']
df_plus['Size'].value_counts()
```

*Not knowing the correct size of the record, this record will also be considered NaN*



```python
df.dtypes
```

```python
df["num_size_float"]= df["num_size"].astype(float) #convert object in float number
```

```python
df.head()
```

*I define the following function*

```python
def string_to_number(row, unit):
    if row[unit] == 'G': #Giga
        return 1000000000
    if row[unit] == 'M': #Mega
        return 1000000
    if row[unit] == 'k': #kilo
        return 1000
    else:
        return np.nan #otherwise
```

```python
df['type_size'] = df.apply(string_to_number, unit='type_size', axis=1)
```

*Let's check the numbers*

```python
df.type_size.value_counts()
```

```python
df.type_size.isnull().sum()
```

*Let's create a new column for the converted data*

```python
df['Size_num'] = df.num_size_float * df.type_size 
df.head()
```

*I delete the columns that are no longer needed*

```python
df.drop(columns=['num_size', 'type_size', 'num_size_float'], inplace = True)
```

```python
df.columns
```
